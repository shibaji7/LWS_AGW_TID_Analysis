```python
# Import all required
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use(["science", "ieee"])

import generate
import wave
```

```python
Amp = [1, .5, 0.25, 0.125, 0.0625, 0.75]
T = [60, 40, 20, 15, 8, 3]
Phi = [10, 20, 30, 40, 50, 60]

data = generate.create_synthetic_data(Amp, Phi, T, 24*60, 60.)
# Plot synthetic dTEC
fig = plt.figure(dpi=240, figsize=(5,2.5))
ax = fig.add_subplot(111)
ax.plot(data["mit"].time/60, data["mit"].dTEC, ls="-", lw=0.2)
ax.set_xlabel("Time, Hours")
ax.set_ylabel("dTEC, TECu")
ax.set_xlim(0, 24)
_ = ax.set_ylim(-3, 3)
```
